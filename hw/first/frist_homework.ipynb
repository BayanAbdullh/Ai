```python
#GET BASE KNOWLEDGE 

import pandas as pd

# تحميل مجموعة البيانات الجديدة
medical_dataset = pd.read_csv('heart.csv')

# عرض مجموعة البيانات
#print(medical_dataset.head())
medical_dataset
```

```python
# عرض القيم الفريدة في عمود الهدف
print(set(medical_dataset['target']))
```

```python
#SPLIT DB  

from sklearn.model_selection import train_test_split

# استخدام الدالة لتقسيم البيانات، الحصول على 3% للاختبار و 97% للتدريب
train_data, test_data = train_test_split(medical_dataset, test_size=0.03, random_state=42)

# طباعة الأشكال (الصفوف × الأعمدة) لمجموعتي البيانات
print("Train data shape:", train_data.shape)
print("Test data shape:", test_data.shape)
```

```python



#CREATE MODEL

from sklearn.tree import DecisionTreeClassifier

# إنشاء نموذج Decision Tree Classifier
model = DecisionTreeClassifier(random_state=1)

# الحصول على عمود الهدف، ليستخدم كهدف للتصنيف
train_diagnoses = train_data['target']

# إسقاط عمود الهدف للحصول على الأعراض
train_patient_symptoms = train_data.drop(columns=['target'])

# بناء شجرة القرار
model.fit(train_patient_symptoms, train_diagnoses)
```

```python

# TEST DATA

from sklearn.metrics import accuracy_score, confusion_matrix

# إسقاط عمود الهدف للحصول على الأعراض للاختبار
test_patient_symptoms = test_data.drop(columns=['target'])

# الحصول على عمود الهدف للاختبار
test_diagnoses = test_data['target']

# التخمين
pred = model.predict(test_patient_symptoms)

# طباعة النتائج
print("Predictions:", pred)

# طباعة دقة النموذج
print("Accuracy Score:", accuracy_score(test_diagnoses, pred))
```

```python

# طباعة مصفوفة الارتباك
print("Confusion Matrix:\n", confusion_matrix(test_diagnoses, pred))
```
